# 04 — Business Insights
Translating KPIs into actionable market intelligence.

```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/processed/brick_features_v4.csv')
district = pd.read_csv('../data/processed/brick_district_summary.csv')
kpis = pd.read_csv('../data/processed/brick_kpi_summary.csv')
print(df.shape, district.shape)
```

```python
print("=== INSIGHT 1: Market Size ===")
print(f"355 active manufacturers across {district['Districts'].nunique()} districts")
print(f"Operational rate: {(df['Status']=='Running').mean()*100:.1f}%")

print("\n=== INSIGHT 2: Eco Transition ===")
print(f"Fly ash share: {(df['Product Type']=='Fly Ash Brick').mean()*100:.1f}%")
print(f"Still clay: {(df['Product Type']=='Fired Clay Brick').mean()*100:.1f}%")

print("\n=== INSIGHT 3: Modernization Gap ===")
print(f"Hand-made: {(df['Machine Type']=='Hand Made').mean()*100:.1f}%")
print(f"Fully automatic: {(df['Machine Type']=='Fully Automatic Machine Made').mean()*100:.1f}%")

print("\n=== INSIGHT 4: Digital Gap ===")
print(f"Has website: {df['has_website'].mean()*100:.1f}%")
print(f"Has contact: {df['has_contact'].mean()*100:.1f}%")

print("\n=== INSIGHT 5: High Risk Segment ===")
high_risk = df[df['is_high_risk']==True]
print(f"High risk factories: {len(high_risk)} ({len(high_risk)/len(df)*100:.1f}%)")
print(f"High risk closure rate: {(high_risk['Status']!='Running').mean()*100:.1f}%")
normal_closure = (df[df['is_high_risk']==False]['Status']!='Running').mean()*100
print(f"Non-high-risk closure rate: {normal_closure:.1f}%")

print("\n=== INSIGHT 6: Geographic Concentration ===")
top5 = district.nlargest(5, 'factory_count')
print(f"Top 5 districts hold {top5['factory_count'].sum()} of 355 factories ({top5['factory_count'].sum()/355*100:.1f}%)")
print(top5[['Districts','factory_count','eco_share','closure_rate']])

print("\n=== INSIGHT 7: The Peddapalli Paradox ===")
peddapalli = district[district['Districts']=='Peddapalli']
print(peddapalli[['Districts','factory_count','eco_share','closure_rate','fully_auto_count']])
```

## Key Business Insights

**Insight 1 — Market is active but transitioning slowly.**
355 factories operate across 31 Telangana districts with an 85.4% operational rate — the market is healthy. However, 34.1% of supply still uses fired clay brick, a product under increasing regulatory pressure. The window for disruption is open but narrowing.

**Insight 2 — Eco transition is urban-led.**
Medchal Malkajgiri (76.2% eco share) and Hyderabad (86.7%) have largely converted to fly ash. Northern districts like Peddapalli (33.3%) and Karimnagar (50.0%) lag significantly. The eco transition follows the highway to Hyderabad — districts further from the metro remain traditional.

**Insight 3 — Modernization is the real gap.**
Only 18% of factories are fully automatic. 55% remain hand-made — the highest-cost, lowest-output production method. A fully automatic fly ash plant entering any district competes against hand-made operations on every dimension: output, consistency, and unit cost.

**Insight 4 — Digital presence is an uncontested moat.**
Only 13.2% of factories have a website. In a market where buyers increasingly search online, a new entrant with even basic digital presence — Google Business profile, website, WhatsApp — operates in a different league from 87% of competitors.

**Insight 5 — High risk factories are not closing faster (yet).**
Hand-made clay factories show a 5.0% closure rate vs 18.5% for other types — counterintuitive. Interpretation: traditional operators are entrenched and slow to exit, but when disruption hits (regulation, input costs), closures will cluster. Early movers capture their customers before the exit wave.

**Insight 6 — Market is concentrated in 5 districts.**
Top 5 districts hold 40% of all factories. Entering a top-5 district means fighting 20-40 established competitors. Districts ranked 6-15 offer 8-16 competitors with similar or better demand fundamentals.

**Insight 7 — The Peddapalli Paradox is the single best opportunity signal.**
Peddapalli sits adjacent to Ramagundam NTPC — one of India's largest thermal power plants and a guaranteed fly ash supply source. Yet only 1 of 21 factories is fully automatic and eco share is just 33.3%. Raw material on the doorstep, modernization gap wide open, closure rate only 4.8% (stable demand). This is the most asymmetric opportunity in the dataset.
