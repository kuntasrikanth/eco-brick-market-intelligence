```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/processed/clean_brick_dataset_v3.csv')
print(df.shape)
```

```python
# Eco flag
df['is_eco'] = df['Product Type'] == 'Fly Ash Brick'

# Operational flag
df['is_operational'] = df['Status'] == 'Running'

# Modernization tier (ordered)
machine_order = {
    'Hand Made': 1,
    'Semi Automatic Concrete Machine Made': 2,
    'Fully Automatic Machine Made': 3,
    'Unknown': 0
}
df['modernization_score'] = df['Machine Type'].map(machine_order)

# Size score (ordered)
size_order = {'Small': 1, 'Moderate': 2, 'Large': 3, 'Unknown': 0}
df['size_score'] = df['Company Size'].map(size_order)

# Risk flag — hand-made clay factories close more (hypothesis to test in ML phase)
df['is_high_risk'] = (
    (df['Machine Type'] == 'Hand Made') & 
    (df['Product Type'] == 'Fired Clay Brick')
)

df.head()
```

```python
print("is_eco distribution:")
print(df['is_eco'].value_counts())
print()
print("modernization_score distribution:")
print(df['modernization_score'].value_counts().sort_index())
print()
print("is_high_risk count:", df['is_high_risk'].sum())
print("High risk as % of market:", round(df['is_high_risk'].mean()*100, 1), "%")
```

```python
df.to_csv('../data/processed/brick_features_v4.csv', index=False)
print(f"Exported v4: {df.shape[0]} rows, {df.shape[1]} columns ✓")
```
